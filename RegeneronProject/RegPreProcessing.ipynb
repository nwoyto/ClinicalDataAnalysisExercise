Adjust heart size for sex and bmi

```python
data_lr = data
```

```python
data_lr['Sex'] = data_lr['Sex'].replace({'m': 0, 'f': 1})
```

### Model adjusts for 

```python
from sklearn.linear_model import LinearRegression

# Extract the predictor variables and outcome variable
X = data_lr[['BMI', 'Age', 'Sex']]
y = data_lr['LVSizePre']

# Create a linear regression object
reg = LinearRegression()

# Fit the model to the data
reg.fit(X, y)

# Print the coefficients of the model
print('Coefficients:', reg.coef_)

```

```python
# Predict left ventricular size for new data
new_data = pd.DataFrame({'BMI': [25.0], 'Age': [50], 'Sex': [0]})
predicted_lvsize = reg.predict(new_data)
print('Predicted left ventricular size:', predicted_lvsize)
```

```python
data_lr.columns
```

```python
#Use the coefficients obtained from the model to predict LVSize2W

data_lr['AdjPreLVSize'] = reg.intercept_ + reg.coef_[0]*data_lr['BMI'] + reg.coef_[1]*data_lr['Age'] + reg.coef_[2]*data_lr['Sex']
```

```python
data_lr.head()
```

```python
#Create a new column and store the quotient of the Values of adjustad and original values

data_lr['AdjCoeff'] = data_lr['AdjPreLVSize'] / data_lr['LVSizePre']
```

```python
# Multiply AdjustmentCoeff by LVSize2W, LVSize4W, and LVSize18M to calculate adjusted values
data_lr['AdjLVSize2W'] = data_lr['LVSize2W'] * data_lr['AdjCoeff']
data_lr['AdjLVSize4W'] = data_lr['LVSize4W'] * data_lr['AdjCoeff']
data_lr['AdjLVSize18M'] = data_lr['LVSize18M'] * data_lr['AdjCoeff']

data_lr['AdjPreLVSize'] = data_lr['AdjPreLVSize'].round(1)
data_lr['AdjLVSize2W'] = data_lr['AdjLVSize2W'].round(1)
data_lr['AdjLVSize4W'] = data_lr['AdjLVSize4W'].round(1)
data_lr['AdjLVSize18M'] = data_lr['AdjLVSize18M'].round(1)
```

```python
data_lr['PatientID'] = data_lr['PatientID'].astype(int)
data_lr['Treatment'] = data_lr['Treatment'].astype(int)
```
